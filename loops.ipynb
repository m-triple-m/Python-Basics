```python
num = 5673
reverse = 0

while num>0:
    print(num, reverse)
    d = num % 10
    reverse = reverse * 10 + d
    num = num // 10
    
print(num, reverse)
```

```python
# WAP to check if a number prime

num2 = int(input('enter a number'))

for i in range(2, num2):
    if num2%i == 0:
        print('not prime')
        break
else:
    print('prime')
```
